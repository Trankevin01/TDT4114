```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return np.sin(3 * np.pi * x) * np.cos(3 * np.pi * y) + 1

# Create a grid of points
x = np.linspace(0, 1/3 , 400)
y = np.linspace(-1/3, 1/3, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the function
plt.figure(figsize=(8, 6))
cp = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(cp)
plt.title('Contour plot of f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Find the global maxima and minima
maxima = np.unravel_index(np.argmax(Z, axis=None), Z.shape)
minima = np.unravel_index(np.argmin(Z, axis=None), Z.shape)

maxima_points = (X[maxima], Y[maxima])
minima_points = (X[minima], Y[minima])

print(f'Global maxima at points: {maxima_points}')
print(f'Global minima at points: {minima_points}')
```
