```python
import numpy as np
import matplotlib.pyplot as plt

# Konstanter
umax = 1.0  # Maks tetthet (biler/m)
vmax = 30.0 # Maks hastighet (m/s)
p_values = [1, 2, 3]    # Ulike p-verdier for v(u)

# Definer funksjonene v(u) og fluksen J(u)
def v(u, p):
    return vmax * (1 - (u/umax)**p)

def J(u, p):
    return u * v(u, p)

# Plot v(u) for ulike p
u_vals = np.linspace(0, umax, 100)
plt.figure(figsize=(8,5))
for p in p_values:
    plt.plot(u_vals, v(u_vals, p), label=f'p={p}')
plt.xlabel('Tetthetet u (biler/m)')
plt.ylabel('Hastighet v (m/s)')
plt.title('Hastighet v(u) for ulike p')
plt.legend()
plt.grid()
plt.show()

# Plott fluksen J(u) for ulike p
plt.figure(figsize=(8, 5))
for p in p_values:
    plt.plot(u_vals, J(u_vals, p), label=f'p={p}')
plt.xlabel('Tetthetet u (biler/m)')
plt.ylabel('FLuks J(biler/s)')
plt.title('Fluks J(u) for ulike p')
plt.legend()
plt.grid()
plt.show()

# Finnee maksimal fluks numerisk
for p in p_values:
    u_max_flux = umax * (1/(p+1))**(1/p)    # Analytisk løsning
    max_flux = J(u_max_flux, p)
    print(f'Maksimal fluks for p={p}: J_max = {max_flux:.2f} biler/s ved u = {u_max_flux:.2f} biler/m')

# Lag to andre vedlig ulike funksjoner v(u)
def v_trig(u):
    return vmax * np.cos((np.pi / 2) * (u / umax))

def v_log(u):
    return vmax * (1 - np.log(1 + u) / np.log(1 + umax))

# Plott de nye funksjonene
plt.figure(figsize=(8, 5))
plt.plot(u_vals, v_trig(u_vals), label='Trigonometrisk v(u)')
plt.plot(u_vals, v_log(u_vals), label='Logaritmisk v(u)')
plt.xlabel('Tetthet u (biler/m)')
plt.ylabel('Hastighet v (m/s)')
plt.title('Alternative hastighetsfunksjoner v(u)')
plt.legend()
plt.grid()
plt.show()
```
